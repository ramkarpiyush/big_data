1. How many types of sets are there in Python?
Python has two types of sets: set (mutable) and frozenset (immutable).

2. What are decorators in Python and how can they be used in daily tasks?
Decorators in Python are functions that modify the behavior of other functions or methods without changing their actual code. They are often used for logging, caching, authentication, and more.

3. How to rank values in descending order in a Python dictionary?
sorted_dict = dict(sorted(my_dict.items(), key=lambda x: x[1], reverse=True))
