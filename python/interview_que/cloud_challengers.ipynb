> Capgemini:
- Input = [2,3,4,5,6]
- Write a python program to square even number and cube odd numbers using
    1. for loop
    2. lambda function
- Expected output:
    - Output = [4, 27, 16, 125, 36]

```python
def square_even_cube_odd_by_forloop(nums:int):
    _output = []
    for i in nums:
        if i%2 == 0:
            _output.append(i**2)
        else:
            _output.append(i**3)
    
    return _output
```

> Theatro Labs:
- string = "aaaasbbbbbcccc"
- Write a python program to count the occurrence of each character in the string
- Expected output:
    - Output: a=4, s=1, b=5, c=4

```python
string = "aaaasbbbbbcccc"
res_dic ={}

for i in string:
    if i in res_dic:
        res_dic[i] += 1
    else:
        res_dic[i]=1

print(res_dic)
```

```python
n1 = 10
n2 = 20
n3 = 15

if (n1>=n2) and (n1>=n3):
    largest = n1
elif (n2>=n1) and (n2>=n3):
    largest = n2
else:
    largest = n3

print(f"Largest Number: {largest}")
```

2. Find the SUM of squares in LIST
    > _list = [1,3,5,7,11,13,17]

```python
_list = [1,3,5,7,11,13,17]

_sum = 0

for i in _list:
    _sum = _sum + (i**2)

print(_sum)
```
