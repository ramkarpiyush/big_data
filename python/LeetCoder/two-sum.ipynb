##### **1. Two Sum:**
- Return indices of the two numbers such that they add up to target

```python
nums = [2, 7, 11, 15]
target = 9
# Output = [0, 1]
```

```python
def get_index_two_nums(nums, target):
    Output = list()
    for i in range(0, len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                Output.extend([i,j])
    return Output

print(get_index_two_nums(nums=nums, target=target))
```

```python
def get_index_two_nums2(nums, target):
    temp = {}
    for i in range(len(nums)):
        j = target-nums[i]
        if j not in temp:
            temp[nums[i]] = i
        else:
            return [temp[j], i]

print(get_index_two_nums2(nums, target))
```

```python
def get_two_sum3(arr, target):
    left = 0
    right = len(arr) - 1

    while left < right:
        current_sum = arr[left] + arr[right]

        if current_sum == target:
            return True
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    return False

print(get_two_sum3(arr=nums, target=target))
```

**2. Three Sum**
- Return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, j != k and nums[i]+nums[j]+nums[k]==0.
- Solutions set must not contains duplicate triplets

```python
nums = [-1, 0, 1, 2, -1, -4]
# Output= [[-1,-1,2], [-1,0,1]]

nums.sort()
out = set()

for i in range(0, len(nums)):
    for j in range(i+1, len(nums)):
        for k in range(j+1, len(nums)):
            if nums[i]+nums[j]+nums[k]==0 :
                out.add((nums[i], nums[j], nums[k]))

# res = list(out)

res = [list(i) for i in out]
print(res)


                
```

```python

def three_sum(nums):
    nums.sort()
    res = []
    n = len(nums)

    for i in range(n):
        # Skip duplicate values for the first number
        if i > 0 and nums[i] == nums[i - 1]:
            continue

        target = -nums[i]
        left, right = i + 1, n - 1

        while left < right:
            two_sum = nums[left] + nums[right]
            if two_sum == target:
                res.append([nums[i], nums[left], nums[right]])

                # Move past duplicates for the second and third numbers
                left_val, right_val = nums[left], nums[right]
                while left < right and nums[left] == left_val:
                    left += 1
                while left < right and nums[right] == right_val:
                    right -= 1

            elif two_sum < target:
                left += 1
            else:
                right -= 1

    return res


# Example usage
questionnums = [-1, 0, 1, 2, -1, -4]
print(three_sum(questionnums))
# Output: [[-1, -1, 2], [-1, 0, 1]]
```
