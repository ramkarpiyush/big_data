# Steps of Binary Search:

- Assume the list is sorted in ascending order:

1. Initialize:

    - low = 0 (start index)

    - high = len(arr) - 1 (end index)

2. Loop until low <= high:

    - Find the middle index:

        > mid = (low + high) // 2

    - Check:

        - If arr[mid] == target: Found, return index.
    
        - If arr[mid] < target: Search right half → low = mid + 1
    
        - If arr[mid] > target: Search left half → high = mid - 1

3. If loop ends, target is not found.

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        
        if arr[mid] == target:
            return mid  # Found → return index
        elif arr[mid] < target:
            low = mid + 1  # Search right half
        else:
            high = mid - 1  # Search left half

    return -1  # Not found
```

```python
numbers = [1, 3, 5, 7, 9, 11, 13]
result = binary_search(numbers, 7)
print(result)  # Output: 3 (index of 7)
```
