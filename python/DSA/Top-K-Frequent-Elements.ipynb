##### **Top K Frequent Elements:**
- **Problem Statement:** You are given an integer array nums and an integer k. Your task is to return the k most frequent elements in the array.

```python
nums = [1,1,1,2,2,3]
k = 2
# Output: [1,2]
```

```python
freq = {}

for i in nums:
    if i not in freq:
        freq[i] = 1
    else:
        freq[i] = freq[i]+1
print(freq)

out = list(filter(lambda x:x[1]>=k, freq.items()))
print(out)

result = [item[0] for item in out]
print(result)
```

```python
sorted_items = sorted(freq.items(), key=lambda x: x[1], reverse=True)

result = [item[0] for item in sorted_items[:k]]
print(result)
```

```python
from collections import Counter

def get_top_K_freq(arr, k):
    n = len(arr)

    mp = Counter(arr)       # Used to count the frequency of elements in an iterable 
                            # (like lists, strings or tuples) or from a mapping (dictionary)

    freq = list(mp.items()) # e.g. [(1, 3), (2, 2), (3, 1)]

    freq.sort(key=lambda x: (x[1], x[0]), reverse=True)
    
    res = []

    for i in range(k):
        res.append(freq[i][0])
        
    return res
```

```python
get_top_K_freq(arr=nums, k=k)
```
