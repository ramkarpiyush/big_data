##### **Merge Intervals**
- **Problem Statement:**
You are given a list of intervals, where each interval is a pair [start, end]. Merge all overlapping intervals and return the list of merged intervals.
- **Approach:** Sorting + Greedy
    1. Sort the intervals based on *start times*.
    2. Initialize merged list.
    3. Iterate through each interval:
        1. If merged is empty or current interval does not overlap with the last merged interval, append it.
        2. Else, merge the current interval with the last one by updating the end time.

```python
intervals = [[1,3], [8,10], [2,6], [15,18]]
```

```python
# Sort intervals by start time
intervals.sort(key=lambda x: x[0])
print(intervals)

# Intervals are sorted based on their start value (x[0]).
```

```python
# Initialize merged list
# start with the first interval as our initial merged interval.
merged = [intervals[0]]
print(merged)
```

```python
from loguru import logger
```

```python
# Iterate through remaining intervals
for current in intervals[1:]:
    last = merged[-1]           # last is the last interval in the merged list, 
                                # which we might need to merge with current
                                
    #Sorted intervals: [[1,3], [2,6], [8,10], [15,18]]
    #
    #merged = [[1,3]]
    #Loop starts at [2,6]:
    #
    #last = [1,3], overlap since 2 <= 3 → update last[1] = max(3,6) = 6
    #merged = [[1,6]]
    #
    #Next [8,10]: 8 <= 6? No → append → merged = [[1,6], [8,10]]
    #Next [15,18]: 15 <= 10? No → append → merged = [[1,6], [8,10], [15,18]]
    
    # Check if intervals overlap
    # If overlapping, we merge them.
    if current[0] <= last[1]:
        last[1] = max(last[1], current[1])
    else:
        merged.append(current)
```

```python
def get_merge_intervals(intervals):
    intervals.sort(key=lambda x: x[0])
    merged = [intervals[0]]

    for current in intervals[1:]:
        last = merged[-1]
        if current[0] <= last[1]:
            last[1] = max(last[1], current[1])
        else:
            merged.append(current)
    return merged

print(get_merge_intervals(intervals))
# Output: [[1, 6], [8, 10], [15, 18]]
```
