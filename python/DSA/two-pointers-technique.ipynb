# Two Pointers Technique:
- The Two-Pointers Technique is a simple yet powerful strategy where you use two indices (pointers) that traverse a data structure—such as an array, list, or string—either toward each other or in the same direction to solve problems more efficiently.


### Naive Method - O(n^2) Time and O(1) Space:
- The very basic approach is to generate all the possible pairs and check if any of them add up to the - target value. 
- To generate all pairs, we simply run two nested loops.

```python
# Sum of Pair Equal to Target
# Given a sorted array arr (sorted in ascending order) and a target, 
# find if there exists any pair of elements (arr[i], arr[j]) such that their sum is equal to the target.
```

```python
# Input: 
arr = [10, 20, 35, 50]
target =70
# Output:  true
# Explanation : There is a pair (20, 50) with given target.
```

```python
def two_sum(arr, target):
    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            if arr[i]+arr[j] == target:
                return True
            
    return False
```

```python
two_sum(arr=arr, target=target)
```

```python
two_sum([10, 15, 20], 50)
```

### Two-Pointer Technique - O(n) time and O(1) space:
- The idea of this technique is to begin with two corners of the given array. 
- We use two index variables left and right to traverse from both corners.

```python
# Initialize: left = 0, right = n - 1
# Run a loop while left < right, do the following inside the loop

# Compute current sum, sum = arr[left] + arr[right]
# If the sum equals the target, we’ve found the pair.
# If the sum is less than the target, move the left pointer to the right to increase the sum.
# If the sum is greater than the target, move the right pointer to the left to decrease the sum.
```

```python
def two_sum(arr, target):
    left = 0 
    right = len(arr) - 1

    while left < right:
        sum = arr[left] + arr[right]

        # Check if the sum matches the target
        if sum == target:
            return True
        elif sum < target: 
            left += 1  # Move left pointer to the right
        else:
            right -= 1 # Move right pointer to the left

    # If no pair is found
    return False

arr = [-3, -1, 0, 1, 2]
target = -2

two_sum(arr, target)
```
