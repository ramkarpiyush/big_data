# Sliding Window Maximum
- Sliding Window Maximum: Deque, Arrays  
- Problem Statement:
    You are given:
    - an integer array nums
    - an integer k (window size)
- Your task is to return the maximum value in each sliding window of size k as the window moves from left to right one step at a time.

````python
# Input:
nums = [1,3,-1,-3,5,3,6,7]
k = 3

# Output: [3, 3, 5, 5, 6, 7]
````

##### Aproch 1:

````python
k_window = []
for i in range(0, len(nums)):
    if i+3 <= len(nums):
        k_window.append(nums[i:i+3])
    else: 
        continue
print(k_window)
````

````python
res = []
for i in k_window:
    i.sort()
    res.append(i[2])
print(res)
````

##### Aproch 2: **O(n) solution using a deque**

````python
from collections import deque

dq = deque()     # Empty deque to stores indices
result = []
````

````python
print(dq)
print(dq[0])
````

````python
from loguru import logger

for i in range(len(nums)):
    # Remove elements out of current window
    if dq and dq[0] == i - k:
        dq.popleft()
    # Remove smaller elements (they can't be max)
    while dq and nums[dq[-1]] < nums[i]:
        dq.pop()
    # Add current index
    dq.append(i)
    # Append max once first window is complete
    if i >= k - 1:
        result.append(nums[dq[0]])
# return result

print(dq)
print(result)
````

````python
# Input:
nums = [1,3,-1,-3,5,3,6,7]
k = 3
# Output: [3, 3, 5, 5, 6, 7]
````

````python
from collections import deque

dq = deque()
res = []

for i in range(len(nums)):
    
    # remove smaller elements
    while dq and nums[dq[-1]] < nums[i]:    # First check: dq
                                            # if dq is True (not empty) then Python evaluates nums[dq[-1]] < nums[i]
                                            # If dq is empty then Python stops immediately
                                            # This is called short-circuit evaluation.
        dq.pop()

    # add current index
    dq.append(i)

    # remove out-of-window index
    if dq[0] == i - k:
        dq.popleft()

    # record max
    if i >= k - 1:
        res.append(nums[dq[0]])

print(res)
````

````python

## ðŸ”¹ i = 0 â†’ nums[0] = 1

### (A) while

```
dq = [] â†’ nothing happens
```

### (B) append

```
dq = [0]
values = [1]
```

### (C) out-of-window?

```
i - k = -3 â†’ no
```

### (D) record?

```
i < k-1 (2) â†’ no
```

---

## ðŸ”¹ i = 1 â†’ nums[1] = 3

### (A) while

```
nums[dq[-1]] = nums[0] = 1
1 < 3 â†’ pop
dq = []
```

### (B) append

```
dq = [1]
values = [3]
```

### (C) out-of-window?

```
i - k = -2 â†’ no
```

### (D) record?

```
i < 2 â†’ no
```

---

## ðŸ”¹ i = 2 â†’ nums[2] = -1

### (A) while

```
nums[1] = 3 < -1 âŒ
```

### (B) append

```
dq = [1, 2]
values = [3, -1]
```

### (C) out-of-window?

```
i - k = -1 â†’ no
```

### (D) record?

```
i >= 2 â†’ YES
res = [ nums[1] ] = [3]
```

âœ… First window `[1,3,-1] â†’ max = 3`

---

## ðŸ”¹ i = 3 â†’ nums[3] = -3

### (A) while

```
nums[2] = -1 < -3 âŒ
```

### (B) append

```
dq = [1,2,3]
values = [3, -1, -3]
```

### (C) out-of-window?

```
i - k = 0
dq[0] = 1 â‰  0 â†’ no removal
```

### (D) record

```
res.append(nums[1]) â†’ 3
res = [3,3]
```

Window `[3,-1,-3] â†’ max = 3`

---

## ðŸ”¹ i = 4 â†’ nums[4] = 5

### (A) while (BIG EVENT ðŸ”¥)

```
nums[3] = -3 < 5 â†’ pop
nums[2] = -1 < 5 â†’ pop
nums[1] = 3  < 5 â†’ pop
dq = []
```

### (B) append

```
dq = [4]
values = [5]
```

### (C) out-of-window?

```
i - k = 1
dq[0] = 4 â‰  1
```

### (D) record

```
res.append(5)
res = [3,3,5]
```

Window `[-1,-3,5] â†’ max = 5`

---

## ðŸ”¹ i = 5 â†’ nums[5] = 3

### (A) while

```
nums[4] = 5 < 3 âŒ
```

### (B) append

```
dq = [4,5]
values = [5,3]
```

### (C) out-of-window?

```
i - k = 2
dq[0] = 4 â‰  2
```

### (D) record

```
res.append(5)
res = [3,3,5,5]
```

Window `[-3,5,3] â†’ max = 5`

---

## ðŸ”¹ i = 6 â†’ nums[6] = 6

### (A) while

```
nums[5] = 3 < 6 â†’ pop
nums[4] = 5 < 6 â†’ pop
dq = []
```

### (B) append

```
dq = [6]
values = [6]
```

### (C) out-of-window?

```
i - k = 3
dq[0] = 6 â‰  3
```

### (D) record

```
res.append(6)
res = [3,3,5,5,6]
```

Window `[5,3,6] â†’ max = 6`

---

## ðŸ”¹ i = 7 â†’ nums[7] = 7

### (A) while

```
nums[6] = 6 < 7 â†’ pop
dq = []
```

### (B) append

```
dq = [7]
values = [7]
```

### (C) out-of-window?

```
i - k = 4
dq[0] = 7 â‰  4
```

### (D) record

```
res.append(7)
res = [3,3,5,5,6,7]
```

Window `[3,6,7] â†’ max = 7`

---

# âœ… FINAL ANSWER

```python
res = [3, 3, 5, 5, 6, 7]
```

---

## FINAL MENTAL MODEL (LOCK THIS ðŸ”’)

For **every element**:

1. Remove smaller elements from back (while)
2. Add current index (append)
3. Remove out-of-window from front
4. Front always gives max

---

If you want next:

* Visual window diagrams
* Why `dq[0]` is always max
* Time complexity proof (O(n))
* Rewrite code in simpler form

Just tell me ðŸ‘
````
