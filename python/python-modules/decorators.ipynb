# **# Decorators:**

```python
def changecase(func):
    def myinner():
        return func().upper()
    return myinner
```

```python
# Without Decorator
def myfunction():
    return "Hello World"
```

```python
print(myfunction())
```

```python
# With Decorator
@changecase
def myfunction2():
    return "Hello World"
```

```python
print(myfunction2())
```

- **Arguments in the Decorated Functions:**
    - Note: Pass the arguments to the wrapper function.

```python
def changecase2(func):
    def myinner(x):
        return func(x).upper()
    return myinner
```

```python
def myfunction(_name):
    return "Mr" + " " + _name

print(myfunction("Piyush Ramkar"))
```

```python
@changecase2
def myfunction(_name):
    return "Mr" + " " + _name

print(myfunction("Piyush Ramkar"))
```

- **Example:** Decorator that runs a function for each item in a list.


```python
def list_iterator(func):
    def inner(nums):
        res = []
        for i in nums:
            res.append(func(i))
        return res
    return inner
```

```python
def get_nums_square(x):
    return x**2

print(get_nums_square([1, 2, 3]))
```

```python
@list_iterator
def get_nums_square(x):
    return x**2

print(get_nums_square([1, 2, 3]))

# inner loops over the list
# func(i) runs the original function (get_nums_square()) on one element
```

- **Example:** Decorator that runs code BEFORE and AFTER a function.


```python
def decorator_name(func):
    def inner_wrapper(x):
        print("Start...")
        res = func(x)
        print("End...")
        return func(x)
    return inner_wrapper

@decorator_name
def get_square(x):
    return x**2

print(get_square(5))
```
