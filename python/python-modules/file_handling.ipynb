# Context Managers:
A context manager in Python is a construct that properly manages resources.
It ensures:
1. Resource is acquired
2. Code block executes
3. Resource is released automatically, even if error occurs

#### Why Context Managers Exist
Without them:
- Files remain open
- Database connections leak
- Memory leaks occur
- Locks are not released
- Transactions remain uncommitted

If working with:

Files → use with open()

Database → use with connection.cursor()

Network → use with

Always prefer context managers.

#### 1. Database:


```python
# Using PostgreSQL with psycopg:
import psycopg

with psycopg.connect(
        host="localhost",
        dbname="postgres",
        user="postgres",
        password="postgres"
    ) as conn:
    with conn.cursor() as cur:
            cur.execute("""
                SELECT table_name
                FROM information_schema.tables
                WHERE table_schema = 'myschema';
            """)

            for row in cur.fetchall():
                print(row[0])            
```

#### How Context Managers Work Internally:
They use two special methods:
1. __enter__()
2. __exit__()


In PostgreSQL using psycopg:

For Connection:
__enter__()

Returns the connection object

Starts a transaction

__exit__()

If no exception → commit()

If exception → rollback()

Closes connection

The with statement internally calls the __enter__() method when entering the block and __exit__() when exiting. In psycopg, __exit__() ensures automatic commit/rollback and resource cleanup, making the code safe and production-ready.

```python
with something as x:

x = something.__enter__()
try:
    # your code
finally:
    something.__exit__()
```

# JSON file handling:

```python
# Read JSON File:

import json

with open('dpt_json.json', 'r') as file:
    df = json.load(file)
    print(df)
```

```python
# Create empty JSON File:

open('newfile.json', 'w').close()
```

```python
# Write (Overwrite) JSON file

import json

data = {"name": "Piyush", "age": 30}

with open("newfile.json", "w") as file:
    json.dump(data, file, indent=4)
    
"""
Note:
# indent=4 tells Python to:
1. Format the JSON output with 4 spaces for each level of indentation
2. Make the file human-readable
"""
```

```python
# Read newfile.json file:

with open("newfile.json", "r") as file:
    res = json.load(file)
    print(res)
```

```python
with open("new_json.json", "x") as file:
    json.dump(data, file, indent=4)

with open("new_json.json", "r") as file:
    res = json.load(file)
    print(res)
```

```python
import json

data = {"name": "Piyush", "age": 30, "weight": 87.50}

try:
    with open("new2.json", "x") as file:
        json.dump(data, file, indent=4)
except FileExistsError:
    with open("new2.json", "r+") as file:
        json.dump(data, file, indent=4)

```
