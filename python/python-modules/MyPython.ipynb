````python
ch = '0'
if ord("A") <= ord(ch) <= ord("Z"):
    print("Uppercase")
elif ord("a") <= ord(ch) <= ord("z"):
    print("Lowercase")
elif ord('0') <= ord(ch) <= ord('9'):
    print("Digit")
else:
    print("Special Character")
````

# Convert List to DataFrame:

```
spark.createDataFrame(data, schema=None)
```

Parameters:
- **data:**	The data to convert. It can be a list of tuples, list of lists, list of dicts, Pandas DataFrame, or RDD.
- **schema** ***(Optional)***: Can be a list of column names, or a StructType schema. If not provided, Spark tries to infer.

````python
nums = [1, 3, 5, 6, 23, 90]

import pandas as pd

df = pd.DataFrame(nums, columns=['col_nm'])
print(df)
````

````python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ListToDataFrame").getOrCreate()

tuple_data = [(i,) for i in nums]
print(tuple_data)

````

````python
df = spark.createDataFrame(tuple_data, ["Value"])
df.show()
````

````python
df2 = spark.createDataFrame(map(lambda x: (x,), nums), ["Value"])
df2.show()
````
