# **Control Flow Statement:**

1. **break:**
    - Immediately terminates the loop (for or while) and jumps to the first statement after the loop.
    - Effect: Loop execution stops at that spot.
    - Use: When you’ve found what you need and don’t want to continue looping.

```python
for i in range(5):
    if i == 3:
        break
    print(i)
print("Loop ended")
```

2. **continue:**
    - Skips the rest of the current iteration and jumps to the next iteration of the loop.
    - Effect: Loop continues, but current iteration ends immediately.
    - Use: When you want to ignore certain cases but keep looping.

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

3. **pass**
    - Does nothing — it’s a placeholder for code you’ll write later or for syntactic completeness.
    - Effect: Execution continues normally after pass.
    - use:
        - Empty function or class body.
        - Placeholder for future logic.
        - To avoid syntax errors when a block is required but you don’t want any action.

```python
for i in range(3):
    if i == 1:
        pass  # placeholder
    print(i)
```
