```python
import psycopg

def pg_conn():
    conn = psycopg.connect(
        host="localhost",
        dbname="postgres",
        user="postgres",
        password="postgres"
    )
    return conn
```

```python
cursor = pg_conn().cursor()

cursor.execute("""
            SELECT table_name
            FROM information_schema.tables
            WHERE table_schema = 'myschema'
               """)

cursor.fetchall()
```

```python
cur = pg_conn().cursor()

cur.execute("""
    SELECT table_name
    FROM information_schema.tables
    WHERE table_schema = 'myschema'
""")

rows = cur.fetchall()

for row in rows:
    print(row[0])

cur.close()
pg_conn().__exit__
```

```python
def get_tables(schema):
    with psycopg.connect(
        host="localhost",
        dbname="postgres",
        user="postgres",
        password="postgres"
    ) as conn:
        with conn.cursor() as cur:
            cur.execute("""
                SELECT table_name
                FROM information_schema.tables
                WHERE table_schema = %s
            """, (schema,))
            return cur.fetchall()

tables = get_tables("myschema")

for table in tables:
    print(table[0])
```
