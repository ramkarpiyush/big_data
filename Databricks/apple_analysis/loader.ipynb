```python
class AbstractLoader:
    def __init__(self, transformedDF):
        self.transformedDF = transformedDF

    def sink(self):
        pass

class AirPodsAfterIphoneLoader(AbstractLoader):
    def sink(self):
        get_sink_source(
            sink_type = "Volume",
            df = self.transformedDF,
            path = "/Volumes/deltacatalog/deltadb/deltavolume/apple_data/airpodsAfterIphone",
            method = "overwrite",
            params = None).load_data_frame()

class OnlyAirPodsAndIphoneLoader(AbstractLoader):
    def sink(self):
        params = {
            "partitionByColumn": ["location"]
        }

        get_sink_source(
            sink_type = "VolumeWithPartition",
            df = self.transformedDF,
            path = "/Volumes/deltacatalog/deltadb/deltavolume/apple_data/OnlyAirpodsAndIphone",
            method = "overwrite",
            params = params).load_data_frame()

        get_sink_source(
            sink_type = "delta",
            df = self.transformedDF,
            path = "deltacatalog.deltadb.OnlyAirpodsAndIphoneDeltaTable",
            method = "overwrite").load_data_frame()
```
