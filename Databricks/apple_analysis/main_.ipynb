```python
import os

folder_path = "/Volumes/deltacatalog/deltadb/deltavolume/dataset/"

for files in os.listdir(folder_path):
    file_format = (files.lower()).split(".")[-1]
    if file_format == "csv":
        spark.read.format("csv").option("header", True).load(folder_path +  f"{files}").show(2)
    elif file_format == "json":
        spark.read.json(folder_path +  f"{files}").show(2)
    elif file_format ==  "parquet":
        spark.read.parquet(folder_path + f"{files}").show(2)
```
