```python
class DataSink():
    """
    Abstarct class
    """
    def __init__(self, df, path, method, params):
        self.df = df
        self.path = path
        self.method = method
        self.params = params or {}

    def load_data_frame(self):
        """
        Abstract Method,
        Function will be defined in sub-classes
        """
        raise ValueError("Not Implemented")

class LoadToVolume(DataSink):   
    def load_data_frame(self):
        self.df.write.mode(self.method).save(self.path)
        print(f"Write data at Location: {self.path}")

class LoadToVolumeWithPartition(DataSink):
    def load_data_frame(self):
        partitionByColumn = self.params.get("partitionByColumn")
        (self.df.write.mode(self.method).partitionBy(partitionByColumn).save(self.path))
        print(f"Write data at Location: {self.path}")

class LoadToVolumeDeltaTable(DataSink):
    def load_data_frame(self):
        (self.df.write.format("delta").mode(self.method).saveAsTable(self.path))
        print(f"Write data at Location: {self.path}")        

def get_sink_source(sink_type, df, path, method, params=None):
    if sink_type == "Volume":
        return LoadToVolume(df, path, method, params)
    elif sink_type == "VolumeWithPartition":
        return LoadToVolumeWithPartition(df, path, method, params)
    elif sink_type == "delta":
        return LoadToVolumeDeltaTable(df, path, method, params)
    else:
        return ValueError(f"Not implemented for sink type: {sink_type}")
    print(f"Sink source inot: {self.sink_type}")
```
