##### **Reading Parquet Files:**
- Reading from parquet:
- Syntax:
```python
spark.read.format("parquet").load("/data/flight-data/parquet/2010-summary.parquet").show()
```

````python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Read-parquet-app")
    .config("spark.hadoop.io.native.lib.available", "false")        # This tells Spark: Donâ€™t use Windows native Hadoop libraries
    .getOrCreate()
)

df = spark.read.format("parquet").load("D:/GitLocal/Spark-The-Definitive-Guide/data/flight-data/parquet/2010-summary.parquet")
df.show()
````
